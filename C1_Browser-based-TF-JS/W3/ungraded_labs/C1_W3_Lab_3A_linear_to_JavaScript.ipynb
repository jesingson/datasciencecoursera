```python
import numpy as np
import tensorflow as tf

print('\u2022 Using TensorFlow Version:', tf.__version__)
```

So let's now try the network. Let's go run for 500 Epochs, but it is very fast because I'm running on a GPU.

```python
model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(units=1, input_shape=[1])  
])

model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)
```

If I tried to predict the value for 10, we'll see it's 18.98 which is great. So now here's the code for me to actually save it.

```python
print(model.predict(np.array([[10.0]], dtype=float)))
```
